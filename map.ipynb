```python
locations_df = pd.read_csv("geocoded_places.csv", header=0, names=['Place', 'Latitude', 'Longitude', 'Address']).drop('Address', axis=1)

# Helper function to create a GeoDataFrame
def create_geodataframe(data, long_col, lat_col, crs='epsg:4326'):
    geometry = [Point(xy) for xy in zip(data[long_col], data[lat_col])]
    return geopandas.GeoDataFrame(data, crs=crs, geometry=geometry)

fig, ax = plt.subplots(figsize=(6, 6))
MAP_GEOJSON:str = f'data/Huesca.geojson'
map = geopandas.read_file(MAP_GEOJSON)
map.plot(ax=ax)
locations_gdf = create_geodataframe(locations_df, 'Longitude', 'Latitude')
locations_gdf.plot(ax=ax, color='red', markersize=50)

```
